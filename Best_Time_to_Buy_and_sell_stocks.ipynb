```python
A=[7,1,5,3,6,4]
```

```python
class Solution:
  def maxProfit(self, A):
   N=len(A)
   left=0 #Buy
   right=1 #sell

   max_profit=0
   while right<N:
     currentProfit=A[right] - A[left]
     if A[left] < A[right]:
       max_profit =max(currentProfit, max_profit)
     else:
       left=right
     right+=1  
       
   return max_profit    
   
```

```python
s=Solution()
s.maxProfit(A)
```
